```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.fft import rfft, rfftfreq
from scipy.signal import butter, filtfilt
```

```python
def generate_sine_wave(freq, sample_rate, duration):
    x = np.linspace(0, duration, sample_rate*duration, endpoint=False)
    frequencies = x*freq
    y = np.sin((2*np.pi)*frequencies)
    return x, y

N = 1024
x = np.sin(2*np.pi / N *np.arange(N))
xi = np.round(3*x)
xQ = 1 / 3 * xi

e = xQ - x

plt.plot(x)
plt.plot(xQ)
plt.plot(e)
plt.show()
```

```python
SAMPLING_RATE = 1000
DURATION = 1
Fa_max = 800

x1, tone1 = generate_sine_wave(Fa_max, SAMPLING_RATE, DURATION)
x2, tone2 = generate_sine_wave(10, SAMPLING_RATE, DURATION)
x3, tone3 = generate_sine_wave(180, SAMPLING_RATE, DURATION)
mixed_signal = tone1+tone2
plt.plot(x1[0:100],tone1[0:100])
plt.show()
```

```python
N = SAMPLING_RATE * DURATION
yf = rfft(tone1)
xf = rfftfreq(N, 1/SAMPLING_RATE)

plt.plot(xf, np.abs(yf))
plt.show()
```
