```python
#  import necessary modules and libraries

import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
# define a sine wave

# define time
t = np.arange(0, 1, 0.01)

# define frequency
f = 2 # Hz

# define amplitude
a = 1


x = np.sin(2*math.pi*2*a*t)
```

```python
# plot the signal x

plt.plot(t,x)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.show()
```

```python
# plot the discrete signal x

plt.plot(t,x, 'o')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.show()
```
